```python
file_nm = 'TEST07_class6' #LAP결과 class 파일
region_class=pd.read_csv(f'{file_nm}.csv',encoding="CP949")
df_class=pd.merge(df,region_class, how='left',on='sgg_cd')
df_output_avg = df_class[['sgg_cd','SIDO_NM_x', 'SGG_NM_x','class'
    ,'popl_tou_rat'            #방문객수/인구수
    ,'visitr_mz_rat'           #전체연령대비 2030 방문자 비율 
    ,'visitr_flower_rat'       #전체연령대비 5060 방문자 비율                  
    ,'dstnc_tou_rat'           #타지역유입비율
    ,'tou_srch_rat'            #검색대비 방문자비율 
    ,'compt_GRDP_rat'    #카드소비액/지역총생산
    ,'AVG_STAY_TM'             #평균체류시간 
    ,'AVG_LODG_DAYS'           #평균숙박일수                 
    ,'tar_num_rat'          #지역별총자원수/107개지역평균
    ,'biz_turs_rat'        #관광사업체 수/전체사업체 수
    ,'sgg_tursm_bdgt'     #SGG전체예산대비_관광예산
    ,'lodg_comp_rat_5','tour_comp_rat_5','shpn_comp_rat_5','sptm_comp_rat_5','food_drk_comp_rat_5'
    , 'fstv_num'              #축제수
    , 'ct_tour_num'           #시티투어수
    , 'socty_corpo_num'       #마을기업수
    , 'INFO_CNT'              #안내소수
    , 'TRAD_COMT'             #문화관광해설사수
    , 'car_pub_zon_num'       #카셰어링존수
    ,'5year_popl'              #5년인구증감률
    ,'popl_dnsty_rat'          #인구밀도
    ,'youth_pur_mvmn_rat'      #청년순이동률
    ,'dtme_popl_num'           #주간인구
    ,'agn_rat'                 #고령화비율
    ,'boyhd_rat'               #유소년비율
    ,'born_rate'               #출생률
    ,'osf_imp_anrev_af_rat'    #재정자립도
    ,'govern_2023'             #재정자주도
    ,'fore_popl_rat'           #외국인비중
    ,'GRDP_rat'                #지역총생산/인구수                      
    ,'sns_mm_avg_cnt'          #시군구별SNS언급량
    ,'train_stp_cnt'	#기차역 수
    ,'bus_stp_cnt'	#버스정류장 수
    ,'subway_stp_cnt'	#지하철 수
    ,'road_pvm_rat'	#도로포장율
    ,'highway_IC_acb'	#고속도로IC접근성
    ,'rail_acb'	#고속/고속화철도 접근성
    ,'park_pop_rat'	#주차장 서비스권역(0.75Km) 내 인구비율
    ,'male_10_rat'
    ,'male_20_rat'                          
    ,'male_30_rat'
    ,'male_40_rat'
    ,'male_50_rat'
    ,'male_60_rat'
    ,'male_70_rat'
    ,'fmale_10_rat'
    ,'fmale_20_rat'                          
    ,'fmale_30_rat'
    ,'fmale_40_rat'
    ,'fmale_50_rat'
    ,'fmale_60_rat'
    ,'fmale_70_rat'
   ]]
#df_output_avg.to_csv('class5_data_list.csv',encoding="CP949" ,index=False)
#df_output_avg5.groupby('class').mean().T.reset_index().to_csv('class5_data_avg.csv',encoding="CP949" ,index=False)
#df_output_avg.groupby('class').mean().T.reset_index().to_csv('class5_data_avg.csv',encoding="CP949" ,index=False)

## 평균 + 평균값 근접 > 지역명 추출

df_output_avg.loc[:,'region_nm'] = df_output_avg.loc[:,'SIDO_NM_x'] + " " +df_output_avg.loc[:,'SGG_NM_x']

class_num = int(file_nm.split('class')[-1])
col_list = list(df_output_avg.iloc[:,4:-1].columns)

df_result_avg = pd.DataFrame(index = col_list)
for i in range(1,class_num+1) :
    df_result_avg[f'class_{i}'] = ''
    for j in range(len(col_list)):
        tmp_df_mean = df_output_avg[df_output_avg['class']==i]

        col_mean_aprx = round((tmp_df_mean[[col_list[j]]]).mean(numeric_only=True),3)[0]
        col_mean_diff = round(abs(tmp_df_mean[[col_list[j]]]-col_mean_aprx).min()[0],3)

        tmp_values = list()
        tmp_values.append('평균 ' + str(col_mean_aprx))
        if round(tmp_df_mean[col_list[j]],3).isin([col_mean_aprx]).sum() >= 1:
            tmp_sgg = list(tmp_df_mean[round(tmp_df_mean[col_list[j]],3).isin([col_mean_aprx])]['region_nm'].reset_index(drop=True))
            tmp_values.append(tmp_sgg)
        elif round(tmp_df_mean[col_list[j]],3).isin([round(col_mean_aprx + col_mean_diff,3)]).sum() >= 1:
            tmp_sgg = list(tmp_df_mean[round(tmp_df_mean[col_list[j]],3).isin([round(col_mean_aprx + col_mean_diff,3)])]['region_nm'])
            tmp_values.append('근삿값 ' + str(round(tmp_df_mean[col_list[j]].mean() + abs(tmp_df_mean[[col_list[j]]]-col_mean_aprx).min()[0],3)))
            tmp_values.append(tmp_sgg)
        elif round(tmp_df_mean[col_list[j]],3).isin([round(col_mean_aprx - col_mean_diff,3)]).sum() >= 1:
            tmp_sgg = list(tmp_df_mean[round(tmp_df_mean[col_list[j]],3).isin([round(col_mean_aprx - col_mean_diff,3)])]['region_nm'])
            tmp_values.append('근삿값 ' + str(round(tmp_df_mean[col_list[j]].mean() - abs(tmp_df_mean[[col_list[j]]]-col_mean_aprx).min()[0],3)))
            tmp_values.append(tmp_sgg)    
        #print(tmp_values)
        df_result_avg.at[col_list[j], f'class_{i}']  = tmp_values
df_result_avg.reset_index(names='column').to_csv(f'{file_nm}_result_mean.csv', encoding="CP949" ,index=False)        
```
